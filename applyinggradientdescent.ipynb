```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
```

```python
data = load_diabetes()
X = data.data[:, 0]
y = data.target

print(X.shape, y.shape)
```

```python
m = 0
b = 0

learning_rate = 0.01

epochs = 1000
n = len(X)

losses = []
```

```python
for i in range(epochs):

    y_pred = m * X + b
    loss = np.mean((y - y_pred)**2)
    losses.append(loss)


    dm = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)


    m = m - learning_rate * dm
    b = b - learning_rate * db

print("Final slope:", m)
print("Final intercept:", b)
```

```python
plt.plot(losses)
plt.xlabel
plt.ylabel
plt.title
plt.show()
```

```python
plt.scatter(X, y, alpha=0.5)
plt.plot(X, m*X + b)
plt.title
plt.show()
```
